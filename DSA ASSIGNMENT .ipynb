# Data visualization

# Scatter plot

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=data,x='PL',y='PW', hue='Classification')
plt.title('PETAL LENGHTH vs PETAL WIDTH')
plt.xlabel('Petal length')
plt.ylabel('petal width')
```

```python
# petal dimensions of the specimens are in the following order
# virginica>versicolor>setosa
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=data,x='SL',y='SW', hue='Classification')
plt.title('SEPAL LENGHTH vs SEPAL WIDTH')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

# Box plot

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=data,x='PL',y='SL',hue='Classification')
plt.figure(figsize=(20,10))
plt.title('PETAL LENGTH VS SEPAL LENGTH')
sns.boxplot(data=data,x='PW',y='SW',hue='Classification')
plt.title('PETAL WIDTH VS SEPAL WIDTH ')
```

```python
## First plot shows linear correlation between petal length and sepal length,virginica having the highest values of both dimensions

## second plot shows exception in sepal width of setosa
## sw of setosa > pw of setosa
## sw of setosa > sw of virginica
```

# Histogram

```python
plt.figure(figsize=(8,8))
plt.hist(data['PL'],rwidth=0.8)
plt.title('PETAL LENGTH')
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['PW'],rwidth=0.8)
plt.title('PETAL WIDTH')
```

# CONCLUSION
